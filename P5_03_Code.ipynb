```python
dataorders2=pd.merge(dataorders, datacustomers, on="customer_id", how = "left", indicator=True)
dataorders2=dataorders2[['order_id','customer_unique_id','order_approved_at']]
dataorders2['order_approved_at']=pd.to_datetime(dataorders2['order_approved_at'])
dataorders2=dataorders2.rename(columns={"customer_unique_id": "customer_id"})
dataorders2
```

```python
exception = dataorders2["_merge"] == "left_only"
dataorders2[exception]
dataorders2 = dataorders2.drop(dataorders2[exception].index, axis=0)
dataorders2=dataorders2[['order_id', 'customer_id', 'payment_value', 'order_approved_at']]
dataorders2.sort_values(by=['order_approved_at'])
```

```python
dataorders2=pd.merge(dataorders2, datareviews, on="order_id", how = "outer", indicator=True)
dataorders2=dataorders2[['order_id', 'customer_id', 'payment_value','review_score', 'order_approved_at']]
dataorders2=dataorders2.dropna(subset=['customer_id', 'order_approved_at'])
dataorders2['order_approved_at']=pd.to_datetime(dataorders2['order_approved_at'])
dataorders2=dataorders2[dataorders2['order_approved_at']  < '2018-09-03']
dataorders2['review_score'] = dataorders2['review_score'].replace(np.nan, 3) 
dataorders2.sort_values(by=['order_approved_at'])
```

```python
datanum=dataordersfinal[['orders_number', 'payment_value', 'review_score','recency']]
datanum['review_score']=datanum['review_score'].astype(float)
datanum
```

```python
def normalize(arr):
    return (arr-arr.mean())/arr.std()
```

```python
datanorm1=normalize(datanum)
datanorm1
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters = 5, init = "k-means++", max_iter = 1000, n_init = 10, random_state = 0)
y_clusters = model.fit_predict(datanorm1)
```

```python
dataordersfinal['clustersm1']=y_clusters
dataordersfinal
```
